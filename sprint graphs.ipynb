```python
import matplotlib.pyplot as plt         #the code plots values of x and y and the list of x and y 
x = ['Sprite','Fanta','Jive','Score'] #brand items and the prices of each item
y = [14,10,10,11]
plt.plot(x,y,'g*--')  #brands graph
#pyplot.title('Cooldrink')
plt.xlabel('Brands')
plt.ylabel('Prices')
plt.show()
```

```python
import matplotlib.pyplot as pyplot   # used plot method to plot line graph, it plots y versus x as lines or makers 
x = ['Ps','KitKat','BarOne','Tex'] 
y = [8,10,13,7]
plt.plot(x,y) #defined the coordinates to be plotted on x and y axes
pyplot.title('Chocolate')
pyplot.xlabel('Chocolate flaours')
pyplot.ylabel('Prices of Chocolate in Rands')
pyplot.show()
```

```python
import matplotlib.pyplot as plt
x = ['Potatoes','Spinach','Onion','Butternut']  #scatter plot defines two lists x and y 
y = [25,15,10,12]
plt.scatter(x,y)
plt.title('Veggies') # the list of veggies and the list of prices
plt.xlabel('Types of Veggies')
plt.ylabel('Price list in Rands')
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x = ['Doritos','Lays','NikNaks','Simba']
y = [15,12,5,18]
plt.bar(x,y, align= 'center')
plt.title('Chips')
plt.ylabel('Prices')
plt.xlabel('Types of chips')
plt.show()
```
